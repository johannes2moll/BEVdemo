```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
IMAGE_H = 469
IMAGE_W = 800

img = cv2.imread('images/1.jpg') # Read the test img
img = img[300:(0+IMAGE_H), 0:IMAGE_W] # Apply np slicing for ROI crop
plt.imshow(img)
```

```python
src = np.float32([[0, IMAGE_H], [800, IMAGE_H], [0, 0], [IMAGE_W, 0]])
dst = np.float32([[330, IMAGE_H], [350, IMAGE_H], [0, 0], [IMAGE_W, 0]])
M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
Minv = cv2.getPerspectiveTransform(dst, src) # Inverse transformation

img = cv2.imread('images/1.jpg') # Read the test img
img = img[300:(0+IMAGE_H), 0:IMAGE_W] # Apply np slicing for ROI crop


warped_img = cv2.warpPerspective(img, M, (IMAGE_W, IMAGE_H)) # Image warping
plt.imshow(cv2.cvtColor(warped_img, cv2.COLOR_BGR2RGB)) # Show results
plt.show()
```
